```python
import pandas as pd
```

```python
data = pd.read_excel("Practical.xlsx")
data
```

```python
# split the total_sales & online_sales
new_data = data["total_sales & online_sales"].str.split("(", expand= True)

# concatinate two df
data = pd.concat([data, new_data], axis = 1)
```

```python
# rename 0 and 1
data.columns = ['food_name', 'date_added', 'data_modified', 'total_sales & online_sales', 'approved', 'total_sales', 'online_sales']

# check if all entries ends with )
data["online_sales"].str.endswith(")").count()

# slice ) in online_sales
data["online_sales"] = data["online_sales"].str.slice(stop = -1)

# drop 'total_sales & online_sales' column
data = data.drop(columns = 'total_sales & online_sales')
data
```
